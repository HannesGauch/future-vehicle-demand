```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data = pd.read_excel("data/transport starter kits/TSDK_ALL.xlsx",sheet_name="Data")

vehicle_stock = data.loc[(data.Variable=="Vehicle stock") & (data["Sub-type"]=="All")& (data["Fuel"]=="All")].groupby("Country name").mean()
motorcycle_stock = data.loc[(data.Variable=="Vehicle stock") & (data["Sub-type"]=="Motorcycle")& (data["Fuel"]=="All")].groupby("Country name").mean()

#clean up for Malaysia:
vehicle_stock.loc["Malaysia",2018]=data.loc[(data.Variable=="Vehicle stock") & (data["Sub-type"]=="All") & (data["Country name"]=="Malaysia")].iloc[0].values[-4]
vehicle_stock.loc["Malaysia",2019]=data.loc[(data.Variable=="Vehicle stock") & (data["Sub-type"]=="All") & (data["Country name"]=="Malaysia")].iloc[0].values[-5]
#clean up for Vietnam:
motorcycle_stock.loc["Viet Nam",2020]=0
# i should take out motorcycles if possible

#subtract motrcycles
for i,row in motorcycle_stock.iterrows():
    print(i)
    print(np.nanmax(vehicle_stock.loc[i,vehicle_stock.columns[1:]]))
    print(max(row[1:]))
    if max(vehicle_stock.loc[i,vehicle_stock.columns[1:]]) > max(row[1:]):
        vehicle_stock.loc[i,vehicle_stock.columns[1:]] = vehicle_stock.loc[i,vehicle_stock.columns[1:]] - row[1:]

#data.loc[(data.Variable=="GDP") & (data["Type"]=="All") & (data["Country name"]=="Taiwan Province of China")].iloc[0]
#data.loc[(data.Variable=="Population") & (data["Type"]=="All")]
#vehicle_stock.index
vehicle_stock
```
