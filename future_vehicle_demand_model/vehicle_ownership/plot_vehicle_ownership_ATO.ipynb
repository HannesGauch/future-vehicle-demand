```python
gdp_data = pd.read_csv("data/gdp/API_NY.GDP.MKTP.KD_DS2_en_csv_v2_5728927.csv",skiprows=4)
pop_data = pd.read_csv("data/population/API_SP.POP.TOTL_DS2_en_csv_v2_5695140/API_SP.POP.TOTL_DS2_en_csv_v2_5695140.csv",skiprows=4)
pop_data.head()
```

```python
fig,ax = plt.subplots()


for i,row in vehicle_data.iterrows():
    #print("try ",i)

    code = row["Economy Code"]

    gdp = np.zeros(len(years))
    pop = np.zeros(len(years))
    try:
        for iy,y in enumerate(years):
            #gdp[iy] = gdp_data.loc[(gdp_data.LOCATION==code)&(gdp_data.TIME==y)]["Value"].values[0]
            #pop[iy] = pop_data.loc[(pop_data.LOCATION==code)&(pop_data.TIME==y)]["Value"].values[0]
            gdp[iy] = gdp_data.loc[(gdp_data["Country Code"]==code)][str(y)].values[0]
            pop[iy] = pop_data.loc[(pop_data["Country Code"]==code)][str(y)].values[0]
        #print("succesfull ",i)

    except:
        print("didn't work ",i,code)
        continue

    
    v_stock = row.values[2:-4]


    ax.plot(gdp/pop/1e3,v_stock/pop*1e3)
    #ax.text(gdp[-5]/1e3/pop[-5], v_stock[-5]/(pop[-5]*1e3), f'{row["Country name"]}')
    #print(gdp[-1],pop[-1],v_stock[-1])
    #ax.set_title(f'{code}')
#ax.set_xscale("log")
#ax.set_yscale("log")
ax.set_xlim([0,100])
ax.set_ylim([0,1100])
ax.set_xlabel("GDP per cap [1000 USD (2015)]")
ax.set_ylabel("vehicle ownership per 1000 capita")
```
