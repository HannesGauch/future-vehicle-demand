```python
def country_map(name):
    if name == "Republic of Moldova":
        return "Moldova, Republic of"
    elif name == "Turkiye":
        return "Turkey"
    elif name=="United Kingdom":
        return "United Kingdom of Great Britain and Northern Ireland"
    elif name == "United States":
        return "United States of America"
    return name
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig,ax = plt.subplots()

for i,row in vehicle_stocks.iterrows():
    #fig,ax = plt.subplots()
    #print("try ",i)
    code = countries.loc[countries.name==country_map(i)]["alpha-3"].values[0]

    gdp = np.zeros(len(years))
    pop = np.zeros(len(years))
    try:
        for iy,y in enumerate(years):
            #gdp[iy] = gdp_data.loc[(gdp_data.LOCATION==code)&(gdp_data.TIME==y)]["Value"].values[0]
            #pop[iy] = pop_data.loc[(pop_data.LOCATION==code)&(pop_data.TIME==y)]["Value"].values[0]
            gdp[iy] = gdp_data2.loc[(gdp_data2["Country Code"]==code)][str(y)].values[0]
            pop[iy] = pop_data2.loc[(pop_data2["Country Code"]==code)][str(y)].values[0]
        #print("succesfull ",i)

    except:
        continue

    
    v_stock = row.values


    ax.plot(gdp/pop/1e3,v_stock/pop*1e3)
    ax.set_title(f'{code}')
    #ax.text(gdp[-5]/1e3/pop[-5], v_stock[-5]/(pop[-5]*1e3), f'{row["Country name"]}')
    #print(gdp[-1],pop[-1],v_stock[-1])
#ax.set_xscale("log")
#ax.set_yscale("log")
ax.set_xlim([0,100])
ax.set_ylim([0,1100])
ax.set_xlabel("GDP per cap [1000 USD (2015)]")
ax.set_ylabel("vehicle ownership per 1000 capita")
```
