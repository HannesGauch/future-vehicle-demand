```python
def map_country_names_UNECE(name_UNECE):
    if name_UNECE == "Republic of Moldova":
        return "Moldova, Republic of"
    elif name_UNECE == "United Kingdom":
        return "United Kingdom of Great Britain and Northern Ireland"
    elif name_UNECE == "United States":
        return "United States of America"
   
    return name_UNECE
```

```python
vehicle_data_europe = pd.read_csv("data/vehicle stock/total-vehicles-in-use-WRS-Europe.csv",index_col=0).T
vehicle_data_europe
```

```python
vehicle_data_oceania = pd.read_csv("data/vehicle stock/total-vehicles-in-use-WRS-Oceania.csv",index_col=0).T
vehicle_data_oceania
```

```python
def map_country_names_WRS(name_WRS):
    if name_WRS == "Bolivia":
        return "Bolivia (Plurinational State of)"
    elif name_WRS == "St. Kitts and Nevis":
        return "Saint Kitts and Nevis"
    elif name_WRS == "St. Lucia":
        return "Saint Lucia"
    elif name_WRS == "St. Vincent and the Grenadines":
        return "St. Vincent and the Grenadines"
    elif name_WRS == "United States":
        return "United States of America"
    elif name_WRS == "Venezuela, RB":
        return "Venezuela (Bolivarian Republic of)"
    elif name_WRS == "Egypt, Arab Rep.":
        return "Egypt"
    elif name_WRS == "SÃ£o TomÃ© and Principe":
        return "Sao Tome and Principe"
    elif name_WRS == "St. Helena":
        return "Saint Helena, Ascension and Tristan da Cunha"
    elif name_WRS == "Tanzania":
        return "Tanzania, United Republic of"
    elif name_WRS == "China, Hong Kong":
        return "Hong Kong"
    elif name_WRS == "China, Macao":
        return "Macao"
    elif name_WRS == "Chinese Taipei":
        return "Taiwan, Province of China"
    elif name_WRS == "Iran, Islamic Rep.":
        return "Iran (Islamic Republic of)"
    elif name_WRS == "Lao PDR":
        return "Lao People's Democratic Republic"
    elif name_WRS == "Occupied Palestinian Territory":
        return "Palestine, State of"
    elif name_WRS == "Vietnam":
        return "Viet Nam"
    elif name_WRS == "Yemen, Rep.":
        return "Yemen"
    elif name_WRS == "Czech Republic":
        return "Czechia"
    elif name_WRS == "Moldova":
        return "Moldova, Republic of"
    elif name_WRS == "Slovak Republic":
        return "Slovakia"
    elif name_WRS == "United Kingdom":
        return "United Kingdom of Great Britain and Northern Ireland"
    elif name_WRS == "Micronesia, Fed. Sts.":
        return "Micronesia (Federated States of)"    

    return name_WRS
```

```python
countries = pd.read_csv("data/countries/all.csv")
countries.columns
vehicle_stocks_con = countries[["name","alpha-3"]].copy().rename(columns={"name":"Country name","alpha-3":"Country code"})
years = np.arange(1990,2023)
for y in years:
    vehicle_stocks_con[y] = np.nan

# data from WRS
years_WRS = np.arange(2015,2021)
for i,row in vehicle_data_americas.iterrows():
    for y in years_WRS:
        if not np.isnan(row[y]):
            vehicle_stocks_con.loc[vehicle_stocks_con["Country name"]==map_country_names_WRS(i),y] = row[y]

for i,row in vehicle_data_oceania.iterrows():
    for y in years_WRS:
        if not np.isnan(row[y]):
            vehicle_stocks_con.loc[vehicle_stocks_con["Country name"]==map_country_names_WRS(i),y] = row[y]

for i,row in vehicle_data_europe.iterrows():
    for y in years_WRS:
        if not np.isnan(row[y]):
            vehicle_stocks_con.loc[vehicle_stocks_con["Country name"]==map_country_names_WRS(i),y] = row[y]


for i,row in vehicle_data_asia.iterrows():
    for y in years_WRS:
        if not np.isnan(row[y]):
            vehicle_stocks_con.loc[vehicle_stocks_con["Country name"]==map_country_names_WRS(i),y] = row[y]

for i,row in vehicle_data_africa.iterrows():
    for y in years_WRS:
        if not np.isnan(row[y]):
            vehicle_stocks_con.loc[vehicle_stocks_con["Country name"]==map_country_names_WRS(i),y] = row[y]

# data from transport data starter kits
years_TDSK = np.arange(1990,2022)
for i,row in vehicle_stock_TDSK.iterrows():
    for y in years_TDSK:
        if not np.isnan(row[y]):
            vehicle_stocks_con.loc[vehicle_stocks_con["Country code"]==row['ISO Alpha-3 code'],y] = row[y]

# data from UNECE
years_UNECE = np.arange(1993,2022)
for i,row in vehicle_stocks_UNECE.iterrows():
    for y in years_UNECE:
        if not np.isnan(row[str(y)]):
            vehicle_stocks_con.loc[vehicle_stocks_con["Country name"]==map_country_names_UNECE(i),y] = row[str(y)]

# data from ATO
years_ATO = np.arange(1990,2023)
for i,row in vehicle_data_ATO.iterrows():
    for y in years_ATO:
        if not np.isnan(row[str(y)]):
            vehicle_stocks_con.loc[vehicle_stocks_con["Country code"]==row['Economy Code'],y] = row[str(y)]




```

```python
vehicle_stocks_con.to_csv("data/vehicle stock/consolidated_vehicle_stock_data.csv")
```
